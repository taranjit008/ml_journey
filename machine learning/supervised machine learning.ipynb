### Machine learning
    machine learning enables computers to learn from data, identify patterns,
    and make decisions with minimal human intervention instead of relying on explicit programming.

<h5> Machine learning algorithms:</h5>
    1: Supervised learning
    2: Unsupervised learning

<h5>Supervised learning:</h5>
   In this process, the algorithm is trained on a dataset containing input-output pairs. The "supervision" comes from the fact that the correct answers (labels) are provided during training. The model iteratively adjusts its internal parameters to minimize the difference between its predictions and the actual labels, effectively learning the underlying relationship between the data features and the target outcome.

<h5>Regression model:</h5>
     This involves predicting a continuous numerical value. Instead of categories, the output is a
     quantity. Common examples include predicting house prices based on features like size and location,
      forecasting stock market trends, or estimating temperature based on weather data.


Ex: House price prediction model

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([400,500,700,900,1100,1300,1500,1550,1600,1800,1880,1900])
y=np.array([180,220,300,340,390,450,490,550,574,600,635,678])
plt.xlabel('Size in SQ. feet')
plt.ylabel('Price in $')

plt.scatter(x,y)
z=np.polyfit(x,y,1)
p= np.poly1d(z)
plt.plot(x,p(x),'r')
plt.show()
```

### Model
\[
f(x)=wx+b
\]

Where:
- x = House size (sq ft)
- y = House price ($1000s)
- w = weight (how much output changes when input changes)
- b = bias (the starting value when the input is 0)

### Cost function
    A cost function in machine learning quantifies the error or discrepancy between a model’s predicted
    values and the actual target values, serving as a metric to evaluate model performance.

    It calculate a value, if cost function value is lower/closer to 0 then it means that the model is
    perfectly fits the data. if its higher then it means model does not fits the data perfectly.

    $$
    J(w,b)=\frac{1}{2m}\sum_{i=1}^{m}\left(\hat{y}^{(i)}-y^{(i)}\right)^2
    $$
